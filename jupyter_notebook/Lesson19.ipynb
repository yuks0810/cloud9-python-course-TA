```python
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tokyo_tower = Image.open("tokyo_tower.png")
print(type(tokyo_tower))
```

```python
plt.imshow(tokyo_tower)
```

```python
gray_tower = tokyo_tower.convert("L")
```

```python
plt.imshow(gray_tower)
```

```python
gray_tower.save("tokyo_tower_gray.png")
```

```python
tokyo_tower_resize = tokyo_tower.resize((250, 375))
plt.imshow(tokyo_tower_resize)
```

```python
tokyo_tower_resize.save("tokyo_tower_resize.png")
```

```python
tokyo_tower_totated = tokyo_tower.rotate(60)
plt.imshow(tokyo_tower_totated)
```

```python
tokyo_tower_translated = tokyo_tower.rotate(0, translate = (100, 100))
plt.imshow(tokyo_tower_translated)
```
